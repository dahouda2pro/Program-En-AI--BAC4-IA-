# Shallow Neural Network Models

# 1. Data Preparation

```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
```

# 2. Loading the Data

```python
data = pd.read_csv('dataset_sdn.csv')
data.head()
```

# 3. EDA

```python
data.info()
```

```python
import seaborn as sns
```

```python
data.drop(['src','dst', 'Protocol'], axis=1, inplace=True)
```

```python
sns.heatmap(data.corr(), annot=True)
```

```python
data.corr().label.sort_values(ascending=False)
```

```python
data = data[['pktcount', 'bytecount', 'pktrate','pktperflow','dur_nsec','switch', 'label']]
```

```python
X = data.drop('label', axis=1)
y = data['label']
```

# 3. Building the Shallow Neural Network 

```python
# Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X
```

```python
y
```

```python
# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Initialize the model
model = Sequential()

# Add the hidden layer with 10 neurons and ReLU activation function
model.add(Dense(10, input_shape=(6,), activation='relu'))

# Add the output layer with sigmoid activation function for binary classification
model.add(Dense(1, activation='sigmoid'))

# Compile the model with Adam optimizer, binary cross-entropy loss, and accuracy metric
model.compile(optimizer=Adam(learning_rate=0.01), loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
# Train the model
history = model.fit(X_train_scaled, y_train, epochs=10, verbose=1, validation_data=(X_test_scaled, y_test))
```

```python
# Evaluate the model on test data
results = model.evaluate(X_test_scaled, y_test)
print(f"Test Loss: {results[0]}, Test Accuracy: {results[1]}")
```
